## Random Forest Improvement with PCA

```python
#Import standardscaler
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
#Remove target columns.
x = red_wine.loc[:,red_wine.columns != 'quality'].values
y = red_wine.loc[:,['quality']].values

#Scale the data
x= pd.DataFrame(StandardScaler().fit_transform(x))
y=pd.DataFrame(y)
# Create PCA object.
pca = PCA(n_components=2)

#Run PCA.
pComp=pca.fit_transform(x)

principalDf = pd.DataFrame(data = pComp
             , columns = ['PC 1', 'PC 2'])

principalDf.head()
```

```python
# Join again the target variable

finalDf = pd.concat([principalDf, red_wine[['quality']]], axis = 1)
finalDf.head()
```

```python
# Show the graphics.
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('PCA', fontsize = 20)
targets = [0.0, 1.0]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['quality'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'PC 1']
               , finalDf.loc[indicesToKeep, 'PC 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)

X_train.shape

# Create PCA object.
pca = PCA(n_components=2)

#Apply PCA on training data
pComp=pca.fit_transform(X_train)

#Run PCA


principalDf = pd.DataFrame(data = pComp
             , columns = ['PC 1', 'PC 2'])

principalDf.head()
```

```python
pcaclf=RandomForestClassifier(n_estimators=10,max_depth=20,random_state=42)

pcaclf.fit(principalDf,y_train.values.ravel())
```

```python
#Apply cross validation to evaluate the results.
scores=cross_val_score(pcaclf,principalDf,y_train.values.ravel(),cv=5)
scores
total_results.append(scores.mean())
total_names.append('PCARandomForestClassifier')
```

```python
#Calculate the mean and the standard deviation of the validation
print("Mean: %0.2f ; Standard Dev.: %0.2f)" % (scores.mean(), scores.std()))
```

# Support Vector Classifier

```python
X = red_wine.drop('quality',axis=1)
y=red_wine['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## SVC Algorithms

```python
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
models = []
models.append(('SupportVectorClassifier', SVC()))
results = []
names = []
scoring = 'accuracy'
for name, model in models:
	kfold = model_selection.KFold(n_splits=10)
	cv_results = model_selection.cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
total_results.append(cv_results.mean())
total_names.append(name)
```

## SVC Improvement

```python
svc= SVC()
svc.fit(X_train,y_train)
pred_svc=svc.predict(X_test)
```

```python
def svc_param_selection(X, y, nfolds):
    param = {
        'C': [0.1, 0.8, 0.9, 1, 1.1, 1.2, 1.3, 1.4],
        'kernel': ['linear', 'rbf'],
        'gamma': [0.1, 0.8, 0.9, 1, 1.1, 1.2, 1.3, 1.4]
    }
    grid_search = GridSearchCV(svc, param_grid=param, scoring='accuracy', cv=nfolds)
    grid_search.fit(X,  y)
    return grid_search.best_params_
```

```python
svc_param_selection(X_train, y_train,10)
```

```python
svc2 = SVC(C = 1.4, gamma =  1.2, kernel= 'rbf')
svc2.fit(X_train, y_train)
pred_svc2 = svc2.predict(X_test)
print('Confusion matrix')
print(confusion_matrix(y_test, pred_svc2))
print('Classification report')
print(classification_report(y_test, pred_svc2))
print('Accuracy score',accuracy_score(y_test, pred_svc2))
total_results.append(accuracy_score(y_test,pred_svc2))
total_names.append('ImprovedSVC')
```

# Method Evaluations

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(mean_absolute_error(test_labels,predictions)))
    print('Accuracy = {:0.2f}%.'.format(accuracy_score(test_labels,predictions)*100))
```

```python
evaluate(svc,X_test,y_test)
evaluate(svc2,X_test,y_test)
evaluate(rfc,X_test,y_test)
```

## Results Discussion

   * Looking at the results of the initial modeling RandomForestClassifier often does marginally worse than the SupportVectorClassifier with normally 2% lower accuracy. The uncertainty comes from the random selection within the test data splitting and the Random Forests. This can lead to the improved version of the Random Forest model to be worse than the unimproved model, but this happens extremely infrequently. Random Forests are often better because they require less optimization after the initial modeling and theoretically platues at 100% after adding about 16 estimators. Though in reality my model peaked at approximately 89.9% only a 0.9% increas after allowing the model to fold and repeat in addition to adding more estimators to the forest. After using parameter optimization for SVC its accuracy was able to rize 6.5% starting at 87.3 coming up to 93.75.
   * The common improvement between both models is using kfolding which allows the model to automatically itterate and essentially train and test multiple times on the training set before finalizing the model. For the RandomForestClassifier I added more estimators in addition to the kfolding and for the SVC I optimized the parameter settings through iteration.
   * The accuracy was tested using cross validation to score.
